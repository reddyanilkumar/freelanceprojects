```python
import pandas as pd
```

```python
data_july15 = pd.read_csv('July15.csv')
data_feb17 = pd.read_csv('feb17.csv')
```

```python
successSignupsJuly15 = data_july15[data_july15['STATUS']==2]['STATUS'].count()
```

```python
successSignupsFeb17 = data_feb17[data_feb17['STATUS']==2]['STATUS'].count()
```

```python
%matplotlib inline   
import matplotlib.pyplot as plt

def customplot(data,value):
    ax = plt.gca()
    y_max = data.max()
    ax.set_ylim([0,y_max*1.2])
    
    for p in ax.patches:
        ax.text(p.get_x() + p.get_width()/2. , p.get_height(),('%d' % int(p.get_height())) if(value == 'count' ) else ('%.2f' % float(p.get_height())),fontsize=12,weight='bold',color='#2874A6',ha='center',va='bottom')
    plt.show()
```

```python
data = pd.DataFrame([{'count':successSignupsJuly15},{'count':successSignupsFeb17}],index=['July_15','Feb_17'])
data.plot(lw=1,kind='line',title='Percentage change of successfully signups',legend=None,color='r')
plt.ylabel('Number of logins')
plt.xlabel('Month_Year')
```

```python
from __future__ import division
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)

x = range(successSignupsFeb17)
y1 = range(10)
y2 = [2*i for i in range(10)]

# plot line graphs
ax.plot(range(successSignupsJuly15), range(successSignupsJuly15))
ax.plot(range(successSignupsFeb17), range(successSignupsFeb17))

# the maximum improvements
x_max = max(x)
y1_max = max(y1)
y2_max = max(y2)
improvements = "{:.2%}".format((y2_max-y1_max)/y1_max) # percentage

ax.arrow(x_max, y1_max, 0, y2_max-y1_max, width=0.1, head_width=0.5, head_length=0.5, color='r', length_includes_head=True)

ax.annotate('100.00', xy=(x_max-2, y2_max-y1_max+3), fontsize=10, color='r')

plt.show()
```
